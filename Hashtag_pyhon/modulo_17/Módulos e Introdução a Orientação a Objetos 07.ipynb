# Mais edições de Gráfico com MatplotLib

```python
#importando o matplotlib
import matplotlib.pyplot as plt
```

- Usando o módulo numpy para trabalhar com números

```python
#importando o módulo numpy
import numpy as np
```

### Outros tipos de Gráfico:

- Linha

```python
vendas = np.random.randint(1000, 3000, 50)
meses = np.arange(1, 51)
```

- Editando o Gráfico de Linha

```python
plt.title('VendasXMeses - Linha')
plt.plot(meses, vendas, '--', color='green')
plt.axis([0, 51, 0, max(vendas)+500])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

```python
#editando o tamanho e a cor da linha - quais parâmetros posso editar?
```

- Dispersão

```python
plt.title('VendasXMeses - Dispersão')
plt.scatter(meses, vendas, color='blue')
plt.axis([0, 51, 0, max(vendas)+500])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

- Barras

```python
plt.title('VendasXMeses - Barras')
plt.bar(meses, vendas, color='red')
plt.axis([0, 51, 0, max(vendas)+500])
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.show()
```

- Trabalhando com Múltiplos Gráficos no mesmo "Plot" -> Para melhor visualização/comparação

```python
plt.figure(1, figsize=(20, 3))

plt.subplot(131)
plt.title('VendasXMeses - Linha')
plt.plot(meses, vendas, '--', color='green')
plt.axis([0, 51, 0, max(vendas)+500])
plt.xlabel('Meses')
plt.ylabel('Vendas')

plt.subplot(132)
plt.title('VendasXMeses - Dispersão')
plt.scatter(meses, vendas, color='blue')
plt.axis([0, 51, 0, max(vendas)+500])
plt.xlabel('Meses')
plt.ylabel('Vendas')

plt.subplot(133)
plt.title('VendasXMeses - Barras')
plt.bar(meses, vendas, color='red')
plt.axis([0, 51, 0, max(vendas)+500])
plt.xlabel('Meses')
plt.ylabel('Vendas')

```
